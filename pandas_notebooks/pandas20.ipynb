Apple Stock

Introduction:

We are going to use Apple's stock price.

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

Step 2. Import the dataset from this address

 Step 3. Assign it to a variable apple

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv"
apple = pd.read_csv(url)
```

```python
apple.head(15)
```

Step 4. Check out the type of the columns


```python
apple.dtypes
```

Step 5. Transform the Date column as a datetime type

```python
apple['Date'] = pd.to_datetime(apple['Date'])
```

```python
apple.head()
```

Step 6. Set the date as the index

```python
apple.set_index('Date', inplace=True)
```

```python
apple.head()
```

Step 7. Is there any duplicate dates?


```python
apple.index.duplicated().any()
```

Step 8. Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple[::-1]
```

Step 9. Get the last business day of each month

```python
apple.resample('BM').last()
```

Step 10. What is the difference in days between the first day and the oldest


```python
difference_in_days = (apple.index[-1] - apple.index[0]).days
difference_in_days
```

Step 11. How many months in the data we have?


```python
num_months = (apple.index[-1].year - apple.index[0].year) * 12 + (apple.index[-1].month - apple.index[0].month)
num_months
```

Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(13.5, 9))
plt.plot(apple['Adj Close'])
plt.title('Apple Stock - Adj Close')
plt.xlabel('Date')
plt.ylabel('Adj Close Value')
plt.grid(True)
plt.show()
```

Step 13. Plot the 'high' and 'low' values on the same row.




```python
import matplotlib.pyplot as plt

# Create a figure and subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Plot High values
axs[0].plot(apple.index, apple['High'], color='blue')
axs[0].set_title('Apple Stock - High')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('High Value')
axs[0].grid(True)

# Plot Low values
axs[1].plot(apple.index, apple['Low'], color='red')
axs[1].set_title('Apple Stock - Low')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Low Value')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
