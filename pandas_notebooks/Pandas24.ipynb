Introduction:

This exercise is a adaptation from the UCI Wine dataset. The only pupose is to practice deleting data with pandas.

Step 1. Import the necessary libraries


```python
import pandas as pd
import numpy as np
```

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called wine

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pd.read_csv(url)
```

```python
wine.head()
```

```python
wine.columns
```

Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns


```python
columns_to_delete = [0, 3, 6, 8, 11, 12, 13]
wine = wine.drop(columns=wine.columns[columns_to_delete])
```

```python
wine.head()
```

Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):

alcohol

malic_acid

alcalinity_of_ash

magnesium

flavanoids

proanthocyanins

hue

```python
wine.columns = ['Alcohol', 'Malic acid','Alcalinity of ash','Magnesium', 'Flavanoids',
                'Proanthocyanins','Hue']
```

```python
wine.head()
```

Step 6. Set the values of the first 3 rows from alcohol as NaN


```python
wine.loc[0:2,'Alcohol']=np.nan
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.head()
```

Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN


```python
wine.loc[2:3 , 'Magnesium'] = np.nan
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.head()
```

Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium


```python
wine['Alcohol'].fillna(10, inplace=True)
wine['Magnesium'].fillna(100, inplace=True)
```

```python
wine.head()
```

Step 9. Count the number of missing values

```python
missing_values = wine.isnull().sum().sum()
print("Number of missing values:", missing_values)
```

Step 10. Create an array of 10 random numbers up until 10

```python
random_numbers = np.random.randint(0, 11, 10)
```

```python
random_numbers
```

Step 11. Use random numbers you generated as an index and assign NaN value to each of cell.


```python
wine.iloc[random_numbers, :] = np.nan
```

```python
wine.head()
```

Step 12. How many missing values do we have?


```python
missing_values_after_random = wine.isnull().sum().sum()
print("Number of missing values after adding random NaNs:", missing_values_after_random)
```

Step 13. Delete the rows that contain missing values

```python
wine.dropna(inplace=True)
```

```python
wine.head()
```

Step 14. Print only the non-null values in alcohol

```python
print("Non-null values in alcohol:")
print(wine['Alcohol'].dropna())
```

Step 15. Reset the index, so it starts with 0 again

```python
wine.reset_index(drop=True, inplace=True)
```
